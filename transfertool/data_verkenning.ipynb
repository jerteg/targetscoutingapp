```python
import pandas as pd

pad = 'C:/Users/Jeroen Tegelaers/Documents/project/transfertool/data/'

transfers = pd.read_csv(pad + 'transfers.csv.gz')
spelers = pd.read_csv(pad + 'players.csv.gz')
clubs = pd.read_csv(pad + 'clubs.csv.gz')
competities = pd.read_csv(pad + 'competitions.csv.gz')

print("=== TRANSFERS ===")
print(transfers.columns.tolist())
print(f"{len(transfers)} rijen")
```

```python
print("=== SPELERS ===")
print(spelers.columns.tolist())
print(f"{len(spelers)} rijen")

print("\n=== CLUBS ===")
print(clubs.columns.tolist())
print(f"{len(clubs)} rijen")

print("\n=== COMPETITIES ===")
print(competities.columns.tolist())
print(f"{len(competities)} rijen")
```

```python
# Hoeveel transfers hebben een echte fee?
betaald = transfers[transfers['transfer_fee'] > 0]
print(f"Totaal transfers: {len(transfers)}")
print(f"Met betaalde fee: {len(betaald)}")
print(f"Percentage: {round(len(betaald)/len(transfers)*100, 1)}%")

# En wat is de datumrange?
transfers['transfer_date'] = pd.to_datetime(transfers['transfer_date'])
print(f"\nVroegste transfer: {transfers['transfer_date'].min()}")
print(f"Laatste transfer: {transfers['transfer_date'].max()}")

# Alleen historisch + betaald
historisch_betaald = transfers[
    (transfers['transfer_fee'] > 0) &
    (transfers['transfer_date'] < '2026-01-01')
]
print(f"\nHistorisch + betaald: {len(historisch_betaald)}")
```

```python
df = transfers[
    (transfers['transfer_fee'] > 0) &
    (transfers['transfer_date'] < '2026-01-01')
].copy()

# Verdeling per seizoen
print("=== TRANSFERS PER SEIZOEN ===")
print(df['transfer_season'].value_counts().sort_index().tail(20))
```

```python
print("=== VERDELING TRANSFER FEES ===")
print(df['transfer_fee'].describe())
print(f"\nOnder €1M:  {len(df[df['transfer_fee'] < 1_000_000])}")
print(f"€1M - €10M: {len(df[(df['transfer_fee'] >= 1_000_000) & (df['transfer_fee'] < 10_000_000)])}")
print(f"€10M - €50M: {len(df[(df['transfer_fee'] >= 10_000_000) & (df['transfer_fee'] < 50_000_000)])}")
print(f"Boven €50M: {len(df[df['transfer_fee'] >= 50_000_000])}")
```

```python
df_met_speler = df.merge(
    spelers[['player_id', 'position', 'sub_position', 'date_of_birth', 'country_of_citizenship']],
    on='player_id',
    how='left'
)

# Bereken leeftijd op moment van transfer
df_met_speler['transfer_date'] = pd.to_datetime(df_met_speler['transfer_date'])
df_met_speler['date_of_birth'] = pd.to_datetime(df_met_speler['date_of_birth'])
df_met_speler['leeftijd_bij_transfer'] = (
    (df_met_speler['transfer_date'] - df_met_speler['date_of_birth']).dt.days / 365.25
).round(1)

print("=== VERDELING PER POSITIE ===")
print(df_met_speler['position'].value_counts())

print("\n=== GEMIDDELDE LEEFTIJD BIJ TRANSFER ===")
print(f"{df_met_speler['leeftijd_bij_transfer'].mean():.1f} jaar")
```

```python
# Filter op transfers vanaf 2018
df_met_speler = df_met_speler[
    df_met_speler['transfer_date'] >= '2018-01-01'
].copy()

print(f"Transfers na 2018: {len(df_met_speler)}")

# Koppel club om competitie te achterhalen
df_volledig = df_met_speler.merge(
    clubs[['club_id', 'domestic_competition_id', 'name']],
    left_on='from_club_id',
    right_on='club_id',
    how='left'
)

# Koppel competitienaam
df_volledig = df_volledig.merge(
    competities[['competition_id', 'name', 'country_name']],
    left_on='domestic_competition_id',
    right_on='competition_id',
    how='left'
)

# Hernoem kolommen voor duidelijkheid
df_volledig = df_volledig.rename(columns={
    'name_x': 'club_naam',
    'name_y': 'competitie_naam',
})

print("\n=== TOP 10 COMPETITIES (aantal uitgaande transfers) ===")
print(df_volledig['competitie_naam'].value_counts().head(10))
```

```python
# Maak leeftijdscategorieën
bins = [15, 18, 21, 24, 27, 30, 33, 40]
labels = ['15-18', '19-21', '22-24', '25-27', '28-30', '31-33', '34+']

df_volledig['leeftijd_categorie'] = pd.cut(
    df_volledig['leeftijd_bij_transfer'],
    bins=bins,
    labels=labels
)

# Bereken gemiddelde en mediaan per cel
basiswaarde = df_volledig.groupby(
    ['competitie_naam', 'position', 'leeftijd_categorie'],
    observed=True
).agg(
    gemiddelde_fee    = ('transfer_fee', 'mean'),
    mediaan_fee       = ('transfer_fee', 'median'),
    aantal_transfers  = ('transfer_fee', 'count')
).reset_index()

# Sorteer op aantal transfers
basiswaarde = basiswaarde.sort_values('aantal_transfers', ascending=False)

print("=== TOP 20 CELLEN (meeste datapunten) ===")
print(basiswaarde.head(20).to_string(index=False))
```

```python
print(f"\nCellen met < 5 transfers: {len(basiswaarde[basiswaarde['aantal_transfers'] < 5])}")
print(f"Cellen met >= 5 transfers: {len(basiswaarde[basiswaarde['aantal_transfers'] >= 5])}")
print(f"Totaal cellen: {len(basiswaarde)}")
```

```python
# Gebruik alleen cellen met >= 5 transfers als betrouwbare basis
betrouwbaar = basiswaarde[basiswaarde['aantal_transfers'] >= 5].copy()

# Voor dunne cellen: val terug op competitie + positie (zonder leeftijd)
fallback = df_volledig.groupby(
    ['competitie_naam', 'position'],
    observed=True
).agg(
    mediaan_fee      = ('transfer_fee', 'median'),
    aantal_transfers = ('transfer_fee', 'count')
).reset_index()

print("=== FALLBACK: MEDIAAN PER COMPETITIE + POSITIE ===")
print(fallback[fallback['competitie_naam'].isin([
    'premier-league', 'bundesliga', 'serie-a', 'ligue-1', 'laliga', 'eredivisie'
])].sort_values('mediaan_fee', ascending=False).to_string(index=False))
```

```python
# Sla beide op als basis voor je model
betrouwbaar.to_csv(
    'C:/Users/Jeroen Tegelaers/Documents/project/transfertool/basiswaarden.csv',
    index=False
)
fallback.to_csv(
    'C:/Users/Jeroen Tegelaers/Documents/project/transfertool/fallback_waarden.csv',
    index=False
)

print("Opgeslagen!")
```

```python
# Bekijk keeper data jupiler pro league als aanvulling
print("=== KEEPERS JUPILER PRO LEAGUE ===")
jupiler_keepers = basiswaarde[
    (basiswaarde['competitie_naam'] == 'jupiler-pro-league') &
    (basiswaarde['position'] == 'Goalkeeper')
][['leeftijd_categorie', 'mediaan_fee', 'aantal_transfers']].to_string(index=False)
print(jupiler_keepers)
```

```python
# Wie zijn de dure Eredivisie aanvallers van 28-30?
uitschieters = df_volledig[
    (df_volledig['competitie_naam'] == 'eredivisie') &
    (df_volledig['position'] == 'Attack') &
    (df_volledig['leeftijd_categorie'] == '28-30')
][['player_name', 'leeftijd_bij_transfer', 'transfer_fee', 'from_club_name', 'to_club_name']].sort_values('transfer_fee', ascending=False)

print(uitschieters.to_string(index=False))
```

```python
# Bereken gemiddelde uitgaande transfer per club
club_multiplier = df_volledig.groupby('from_club_name')['transfer_fee'].agg([
    'median', 'count'
]).reset_index()
club_multiplier.columns = ['club', 'mediaan_fee', 'aantal_transfers']
club_multiplier['competitie_mediaan'] = club_multiplier['mediaan_fee'].median()
club_multiplier['club_multiplier'] = (
    club_multiplier['mediaan_fee'] / club_multiplier['competitie_mediaan']
).round(2)

# Toon Eredivisie clubs
eredivisie_clubs = ['Ajax', 'PSV', 'Feyenoord', 'AZ', 'Utrecht', 
                    'Vitesse', 'Twente', 'Heerenveen', 'Groningen']

print(club_multiplier[
    club_multiplier['club'].isin(eredivisie_clubs)
].sort_values('club_multiplier', ascending=False).to_string(index=False))
```

```python
# Bereken mediaan per club + positie + leeftijdscategorie
club_pos_leeftijd = df_volledig[
    df_volledig['competitie_naam'] == 'eredivisie'
].groupby(
    ['from_club_name', 'position', 'leeftijd_categorie'],
    observed=True
).agg(
    club_mediaan = ('transfer_fee', 'median'),
    aantal       = ('transfer_fee', 'count')
).reset_index()

# Voeg competitie mediaan per positie + leeftijd toe
club_pos_leeftijd = club_pos_leeftijd.merge(
    eredivisie_gesorteerd[['position', 'leeftijd_categorie', 'mediaan_fee']],
    on=['position', 'leeftijd_categorie'],
    how='left'
)

# Bereken club multiplier per cel
club_pos_leeftijd['club_multiplier'] = (
    club_pos_leeftijd['club_mediaan'] / club_pos_leeftijd['mediaan_fee']
).round(2)

# Toon Ajax
print("=== AJAX CLUBMULTIPLIER PER CEL ===")
print(club_pos_leeftijd[
    (club_pos_leeftijd['from_club_name'] == 'Ajax') &
    (club_pos_leeftijd['aantal'] >= 3)
].sort_values('club_multiplier', ascending=False).to_string(index=False))
```

```python
# Hoeveel transfers heeft elke Eredivisie club totaal?
print("=== AANTAL TRANSFERS PER EREDIVISIE CLUB ===")
eredivisie_clubs_data = df_volledig[
    df_volledig['competitie_naam'] == 'eredivisie'
].groupby('from_club_name').agg(
    aantal_transfers = ('transfer_fee', 'count'),
    mediaan_fee      = ('transfer_fee', 'median'),
    totale_inkomsten = ('transfer_fee', 'sum')
).reset_index().sort_values('aantal_transfers', ascending=False)

print(eredivisie_clubs_data.head(15).to_string(index=False))
```
